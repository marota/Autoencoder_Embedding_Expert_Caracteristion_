```python
import numpy as np, pandas as pd
import time
from datetime import datetime
```

```python
stations = pd.read_csv("postesSynop.csv", sep=";", usecols=[0,1,2,3], dtype=str, nrows=42) #.drop(axis=0, index=[5])
stations['ID'] = stations.ID.astype(int)
stations
```

```python
weights = pd.read_csv('WeatherStationsRTE.csv', sep=';')
weights
```

```python
test_2016 = pd.read_csv("synop.201607.csv", sep=';')[['date','numer_sta','t']]
test_2016['date']=[datetime(year = int(str(dt)[:4]), month=int(str(dt)[4:6]), day=int(str(dt)[6:8]), hour=int(str(dt)[8:10])) for dt in test_2016.date]
id_sta = [i for i in range(test_2016.shape[0]) if test_2016.numer_sta.iloc[i] in np.unique(stations.ID.values)]
test_2016 = test_2016.iloc[id_sta].reset_index()
test_2016['t'].iloc[np.where(test_2016['t']=='mq')]='Nan'
test_2016['t'] = test_2016.t.astype(float) - 273.15
test_2016.info()
```

```python
list_csv = ["synop.20180{}.csv" .format(int(i+1)) for i in range(9)] + ["synop.20181{}.csv" .format(int(i)) for i in range(3)] 
```

```python
list_frame=[]
for csv in list_csv:
    test = pd.read_csv(csv, sep=';')[['date','numer_sta','t']]
    test['date']=[datetime(year = int(str(dt)[:4]), month=int(str(dt)[4:6]), day=int(str(dt)[6:8]), hour=int(str(dt)[8:10])) for dt in test.date]
    id_sta = [i for i in range(test.shape[0]) if test.numer_sta.iloc[i] in np.unique(stations.ID.values)]
    test = test.iloc[id_sta].reset_index()
    test['t'].iloc[np.where(test['t']=='mq')]='Nan'
    test['t'] = test.t.astype(float) - 273.15
    list_frame.append(test)
```

```python
temp_2018 = pd.concat(list_frame, axis=0, ignore_index=True).reset_index().drop(labels=['level_0', 'index'], axis=1)
```

```python
list_stations = stations.ID.values
```

```python
temperature_2016 = pd.pivot_table(test_2016, values='t', index='date', columns='numer_sta')[list_stations]
```

```python
temperature_2018 = pd.pivot_table(temp_2018, values='t', index='date', columns='numer_sta')[list_stations]
```

```python
temperature_2016.isna().sum(axis=0)
```

```python
threshold = 20
sta_na =list_stations[temperature_2018.isna().sum(axis=0)>threshold]
sta_na
```

```python
temperature_2016 = temperature_2016.drop(columns=sta_na)
temperature_2018 = temperature_2018.drop(columns=sta_na)
```

```python
temperature_2016 = temperature_2016.interpolate(method='linear')
temperature_2018 = temperature_2018.interpolate(method='linear')
```

```python
dataset=pd.read_csv("dataset.csv")[['ds', 'temperature_France']]
dataset['ds']=dataset.ds.astype('datetime64')
```

```python
id_test = np.asarray([int(np.where(dataset.ds == temperature_2016.index[k])[0]) for k in range(temperature_2016.shape[0])])
```

```python
temp_France = pd.DataFrame(data=dataset.temperature_France.iloc[id_test].values, columns=['t0'])
temp_France['t0+30min'] = dataset.temperature_France.iloc[id_test+1].values
temp_France['t0+1h'] = dataset.temperature_France.iloc[id_test+2].values
temp_France['t0+1h30'] = dataset.temperature_France.iloc[id_test+3].values
temp_France['t0+2h'] = dataset.temperature_France.iloc[id_test+4].values
temp_France['t0+2h30'] = dataset.temperature_France.iloc[id_test+5].values
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators=100, max_depth=12, random_state=42)
```

```python
reg.fit(temperature_2016.values, temp_France['t0'].values)
```

```python
t_pred = reg.predict(temperature_2016.values)
np.sqrt(np.sum((t_pred - temp_France['t0'].values)**2))
```

```python
reg.feature_importances_
```

```python
from xgboost import XGBRegressor
reg_xgb = XGBRegressor(n_estimators=200, learning_rate=0.05, max_depth=3)
```

```python
reg_xgb.fit(temperature_2016.values, temp_France['t0'].values)
t_pred_xgb = reg_xgb.predict(temperature_2016.values)
np.sqrt(np.sum((t_pred_xgb - temp_France['t0'].values)**2))
```

```python
np.sqrt(np.sum(((0.99*t_pred_xgb+0.01*t_pred) - temp_France['t0'].values)**2))
```
